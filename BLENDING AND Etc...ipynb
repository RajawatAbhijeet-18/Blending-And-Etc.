```python
myimg = np.zeros((512,512,3), dtype=np.uint8 )
```

```python
myimg_rev = cv2.bitwise_not(myimg)
```

```python
myimg2 = np.zeros((512,512,3), dtype=np.uint8 )
```

```python
myimg2 = cv2.rectangle(myimg2 , (100,50) ,(400,300) ,(255,255,255),-1)
```

```python
plt.imshow(myimg2)
```

```python
img_add = cv2.bitwise_and(myimg , myimg2)
```

```python
plt.imshow(img_add)
```

```python
img_or = cv2.bitwise_or(myimg , myimg2)
```

```python
plt.imshow(img_or)
```
